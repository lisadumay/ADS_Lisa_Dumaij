```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

T0 = pd.read_csv('./T0_data2.csv', sep=';')

print(T0['BMI'])

#er mogen geen dubbele ID tussen zitten. Dit zou betekenen dat een kind meerdere keer gemeten wordt.
T0.duplicated(subset = 'Respondentnummer', keep = 'last')

# Vervolgens heb ik de data van de T0 in een dataframe gezet, zodat ik de volgende stappen kan uitvoeren
df = pd.DataFrame(data=T0)

# De datum wordt telkens anders vermeld
df['Geboortedatum'] = df['Geboortedatum'].str.replace('-','/')
df['Geboortedatum'] = df['Geboortedatum'].str.replace('.','/')



# Alle ratings met nul weg in de BMI, omdat nul ratings aangeeft dat er geen rating heeft plaatsgevonden.
df = df[df.BMI != ""]

print(df)

```

```python
sns.stripplot('MQ', 'Lengte (in m)', hue='MQ catogory', data=df, jitter=True , size =2)
plt.title('MQ in compersiment to the Weight')
plt.xlabel('MQ score')
plt.ylabel('Wight in KG')
plt.ylim(0,300)
plt.xlim(1,5)
plt.ion()
plt.show()
```

```python
sns.lmplot(x='MQ',y='BMI',data= df, hue='MQ catogory')
plt.title('the MQ score in contest to the BMI sorted by category')
plt.xlabel('MQ score')
plt.ylabel('BMI score')
plt.xlim(0,200)
plt.ylim(0,30)
plt.ion()
plt.show()
```

```python
plt.hist(df['BMI'], color='skyblue', label='alle versies', bins = [0,5,10,15,20,25,30],rwidth=0.7)
plt.title('BMI')
plt.xlabel('Cijfer')
plt.ylabel('Frequentie')
plt.xlim(1,30)
plt.ion()
plt.show()
```

```python
plt.scatter(df['MQ'], df['BMI'], color='pink')
plt.title('Scatter Plot relation BMI and MQ')
plt.xlabel ('MQ score')
plt.ylabel('BMI score')
plt.show()
```

```python
sns.barplot(df['MQ catogory'], df['Lengte (in m)'], color ='MediumSeaGreen')
plt.title('barplot Plot relation BMI and MQ')
plt.xlabel ('MQ score')
plt.ylabel('length (m)')
plt.show()
```

```python
sns.boxplot(x='Lengte (in m)', data=df, color='Steelblue' )
plt.title('Boxplot lenght')
plt.xlabel ('length childern')
plt.xlim(0.8,1.75)
plt.show()
```

```python
sns.boxplot(x='Gender', y= 'Lengte (in m)', data=df, color='Steelblue' )
plt.title('Boxplot length divided by Gender')
plt.xlabel('Gender')
plt.ylabel ('length childern')
plt.ylim(0.8,1.75)
plt.show()
```

```python
sns.distplot(df['BMI'],kde=True, bins=[5,7.5,10,12.5,15,17.5,20,22.5,25,27.5,30])
plt.xlim(5,30)

plt.show()
```

```python
plt.plot('gewicht (afgerond op 0.1 kg)', 'Lengte (in m)', data=df, kind='scatter')
```

```python
sns.countplot(x='MQ catogory', data=df, color='Orange')
plt.title('Countplot MQ score')
plt.xlabel ('MQ score')
plt.ylabel('frequency')
plt.show()

%%javascript
Jupyter.notebook.session.delete()
```
